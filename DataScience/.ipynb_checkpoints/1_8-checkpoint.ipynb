```python
my_stat["V5"] = my_stat['V1'] + my_stat['V4']
my_stat["V6"] = np.log(my_stat["V2"])
```

```python
my_stat = my_stat.rename(columns = {"V1":"session_value", "V2" : "group", "V3":"time", "V4": "n_users"})
```

```python
df_stats = pd.read_csv("my_stat_1.csv")
my_stat["session_value"] = my_stat["session_value"].fillna(0)
median = my_stat[ (my_stat["n_users"]>=0) & (my_stat['n_users'] != -0.0)]["n_users"].median()
my_stat.loc[my_stat['n_users']< 0.0, 'n_users'] = median
my_stat["n_users"] = my_stat["n_users"].fillna(median)
```
